```python
import matplotlib.pyplot as plt
import numpy as np
import cv2
from collections import deque
import datetime


# Define the upper and lower boundaries for a color to be considered "Blue" -  by 'HSV'
blueLower = np.array([100, 120, 60])
blueUpper = np.array([140, 255, 255])

# Define a 5x5 kernel for erosion and dilation
kernel = np.ones((5, 5), np.uint8)

# Setup deques to store separate colors in separate arrays
bpoints = [deque(maxlen=512)]
gpoints = [deque(maxlen=512)]
rpoints = [deque(maxlen=512)]
ypoints = [deque(maxlen=512)]

bindex = 0
gindex = 0
rindex = 0
yindex = 0

lastIndex = 0


by=bz=gx=gz=rx=ry=yx= 0
bx = 255
gy = 255
rz = 255
yy = yz = 255

color3 = [255,255,255,255]

colorIndex = 0

# Load the video
camera = cv2.VideoCapture(0)

contour_scale=10

counter=0

#All 3 X(i) are for the '*' place
x1=158
x2=318
x3=478
x4=638

#For the saving option
flagt=0

#For the counter
flagc=0
flagc_up=0
sec58up=100

# Keep on looping
while True:

    bx = color3[0]
    gy = color3[1]
    rz = color3[2]
    yy = yz = color3[3]
    colors = [(bx, by, bz) , (gx, gy, gz) , (rx, ry, rz) , (yx, yy, yz)]
    
    # Setup the Paint interface
    paintWindow = np.zeros((480,640,3)) + 255 #plain white
    
    
    #### Rectangle ####
    paintWindow = cv2.rectangle(paintWindow, (2,1), (158,55), (bx,0,0), 4) #B
    paintWindow = cv2.rectangle(paintWindow, (162,1), (318,55), (0,gy,0), 4) #G
    paintWindow = cv2.rectangle(paintWindow, (322,1), (478,55), (0,0,rz), 4) #R
    paintWindow = cv2.rectangle(paintWindow, (482,1), (638,55), (0,yy,yz), 4) #Y
    #
    paintWindow = cv2.rectangle(paintWindow, (0,420), (150,470), (0, 0, 0), 4) #Save
    paintWindow = cv2.rectangle(paintWindow, (490,420), (640,470), (0, 0, 0), 4) #Clear
       
    #### TEXT - on paintWindow ####
    cv2.putText(paintWindow, "BLUE", (20, 25), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 0, 0), 2, cv2.LINE_AA)
    cv2.putText(paintWindow, "GREEN", (180, 25), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 0, 0), 2, cv2.LINE_AA)
    cv2.putText(paintWindow, "RED", (340, 25), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 0, 0), 2, cv2.LINE_AA)
    cv2.putText(paintWindow, "YELLOW", (500, 25), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 0, 0), 2, cv2.LINE_AA)
    #
    cv2.putText(paintWindow, "SAVE", (10, 450), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 0, 0), 2, cv2.LINE_AA)
    cv2.putText(paintWindow, "CLEAR", (500, 450), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 0, 0), 2, cv2.LINE_AA)
    #
    cv2.putText(paintWindow, "*", (x1-12, 70), cv2.FONT_HERSHEY_SIMPLEX, 0.5, colors[0], 2, cv2.LINE_AA)
    cv2.putText(paintWindow, "*", (x2-12, 70), cv2.FONT_HERSHEY_SIMPLEX, 0.5, colors[1], 2, cv2.LINE_AA)
    cv2.putText(paintWindow, "*", (x3-12, 70), cv2.FONT_HERSHEY_SIMPLEX, 0.5, colors[2], 2, cv2.LINE_AA)
    cv2.putText(paintWindow, "*", (x4-12, 70), cv2.FONT_HERSHEY_SIMPLEX, 0.5, colors[3], 2, cv2.LINE_AA)
    #
    cv2.putText(paintWindow, "{0}".format(x1+97), (x1-48, 85), cv2.FONT_HERSHEY_SIMPLEX, 0.5, colors[0], 1, cv2.LINE_AA)
    cv2.putText(paintWindow, "{0}".format(x2+97-160), (x2-48, 85), cv2.FONT_HERSHEY_SIMPLEX, 0.5, colors[1], 1, cv2.LINE_AA)
    cv2.putText(paintWindow, "{0}".format(x3+97-320), (x3-48, 85), cv2.FONT_HERSHEY_SIMPLEX, 0.5, colors[2], 1, cv2.LINE_AA)
    cv2.putText(paintWindow, "{0}".format(x4+97-480), (x4-48, 85), cv2.FONT_HERSHEY_SIMPLEX, 0.5, colors[3], 1, cv2.LINE_AA)

  
    # Retrive the current paintWindow
    (ret, frame) = camera.read()
    frame = cv2.flip(frame, 1)
    frame=cv2.resize(frame,(640,480))
    hsv = cv2.cvtColor(frame, cv2.COLOR_BGR2HSV)
    
 
    # Showing instructions for work
    if(counter<280):
        if (counter<250):
            cv2.putText(paintWindow, "1) Drag the '*' to change the color value", (10, 120), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0,0,0), 1, cv2.LINE_AA)
            cv2.putText(frame, "1) Drag the '*' to change the color value", (10, 120), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0,0,0), 1, cv2.LINE_AA)
        if(counter>50 and counter<260):
            cv2.putText(paintWindow, "2) Press:  's' for Save    |    'c' for Clear    |    'q' / 'Esc' for Quit", (10, 150), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0,0,0), 1, cv2.LINE_AA)
            cv2.putText(frame, "2) Press:  's' for Save    |    'c' for Clear    |    'q' / 'Esc' for Quit", (10, 150), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0,0,0), 1, cv2.LINE_AA)
        if(counter>100 and counter<270):
            cv2.putText(paintWindow, "3) Press: 'b' for Blue | 'g' for Green | 'r' for red | 'y' for yellow", (10, 180), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0,0,0), 1, cv2.LINE_AA)
            cv2.putText(frame, "3) Press: 'b' for Blue | 'g' for Green | 'r' for red | 'y' for yellow", (10, 180), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0,0,0), 1, cv2.LINE_AA)
        if(counter>150):
            cv2.putText(paintWindow, "Enjoy...", (20, 280-int(counter*0.2)), cv2.FONT_HERSHEY_SIMPLEX, 1.5, (0,0,0), 3, cv2.LINE_AA)
            cv2.putText(frame, "Enjoy...", (20, 280-int(counter*0.2)), cv2.FONT_HERSHEY_SIMPLEX, 1.5, (0,0,0), 3, cv2.LINE_AA)
        #speed of instructions
        counter=counter+2
    
    
    paintWindow = cv2.rectangle(paintWindow, (2,100), (640,380), (0,0,0), 2) 

       
    #### Rectangle ####
    frame = cv2.rectangle(frame, (2,1), (158,55), (bx,0,0), 4) #B
    frame = cv2.rectangle(frame, (162,1), (318,55), (0,gy,0), 4) #G
    frame = cv2.rectangle(frame, (322,1), (478,55), (0,0,rz), 4) #R
    frame = cv2.rectangle(frame, (482,1), (638,55), (0,yy,yz), 4) #Y
    #
    frame = cv2.rectangle(frame, (0,420), (150,470), (255, 255, 255), 4) #Save
    frame = cv2.rectangle(frame, (490,420), (640,470), (255, 255, 255), 4) #Clear
     
    #### TEXT - on frame ####
    cv2.putText(frame, "BLUE", (20, 25), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (255, 255, 255), 2, cv2.LINE_AA)
    cv2.putText(frame, "GREEN", (180, 25), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (255, 255, 255), 2, cv2.LINE_AA)
    cv2.putText(frame, "RED", (340, 25), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (255, 255, 255), 2, cv2.LINE_AA)
    cv2.putText(frame, "YELLOW", (500, 25), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (255, 255, 255), 2, cv2.LINE_AA)
    #
    cv2.putText(frame, "SAVE", (10, 450), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (255, 255, 255), 2, cv2.LINE_AA)
    cv2.putText(frame, "CLEAR", (500, 450), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (255, 255, 255), 2, cv2.LINE_AA)
    #
    cv2.putText(frame, "*", (x1-12, 70), cv2.FONT_HERSHEY_SIMPLEX, 0.5, colors[0], 2, cv2.LINE_AA)
    cv2.putText(frame, "*", (x2-12, 70), cv2.FONT_HERSHEY_SIMPLEX, 0.5, colors[1], 2, cv2.LINE_AA)
    cv2.putText(frame, "*", (x3-12, 70), cv2.FONT_HERSHEY_SIMPLEX, 0.5, colors[2], 2, cv2.LINE_AA)
    cv2.putText(frame, "*", (x4-12, 70), cv2.FONT_HERSHEY_SIMPLEX, 0.5, colors[3], 2, cv2.LINE_AA)
    #
    cv2.putText(frame, "{0}".format(x1+97), (x1-48, 85), cv2.FONT_HERSHEY_SIMPLEX, 0.5, colors[0], 1, cv2.LINE_AA)
    cv2.putText(frame, "{0}".format(x2+97-160), (x2-48, 85), cv2.FONT_HERSHEY_SIMPLEX, 0.5, colors[1], 1, cv2.LINE_AA)
    cv2.putText(frame, "{0}".format(x3+97-320), (x3-48, 85), cv2.FONT_HERSHEY_SIMPLEX, 0.5, colors[2], 1, cv2.LINE_AA)
    cv2.putText(frame, "{0}".format(x4+97-480), (x4-48, 85), cv2.FONT_HERSHEY_SIMPLEX, 0.5, colors[3], 1, cv2.LINE_AA)

    
 
    # Check to see if we have reached the end of the video
    if not ret:
        break

    # Determine which pixels fall within the blue boundaries and then blur the binary image
    blueMask = cv2.inRange(hsv, blueLower, blueUpper)
    blueMask = cv2.erode(blueMask, kernel, iterations=2)
    blueMask = cv2.morphologyEx(blueMask, cv2.MORPH_OPEN, kernel)
    blueMask = cv2.dilate(blueMask, kernel, iterations=1)

    
    # Find contours in the image
    (_, cnts, _) = cv2.findContours(blueMask.copy(), cv2.RETR_EXTERNAL,
    	cv2.CHAIN_APPROX_SIMPLE)
    center = None

    # Check to see if any contours were found
    if len(cnts) > 0:
    	# Sort the contours and find the largest one
        cnt = sorted(cnts, key = cv2.contourArea, reverse = True)[0]
        # Get the radius of the enclosing circle around the found contour
        ((x, y), radius) = cv2.minEnclosingCircle(cnt)
        
        # Draw the circle around the contour
        if (int(radius) > contour_scale):
            cv2.circle(frame, (int(x), int(y)), int(radius*1.2),  colors[colorIndex], 2)
            cv2.circle(paintWindow, (int(x), int(y)), int(radius*1.2),  colors[colorIndex], 2)
        else:   
            cv2.circle(frame, (int(x), int(y)), int(radius*1.2), (0, 0, 0), 2)
            cv2.circle(frame, (int(x), int(y)), int(radius*1.8),  colors[colorIndex], 2)
            #
            cv2.circle(paintWindow, (int(x), int(y)), int(radius*1.2), (0, 0, 0), 2)
            cv2.circle(paintWindow, (int(x), int(y)), int(radius*1.8),  colors[colorIndex], 2)

            
        # Get the moments to calculate the center of the contour (in this case Circle)
        M = cv2.moments(cnt)
        center = (int(M['m10'] / M['m00']), int(M['m01'] / M['m00']))
        
        #check if the center of the paintbrush touch on of the colors
        
        #Upper section
        if center[1] <= 118:
            if center[1] <= 85:
                if 2 <= center[0] <= 158: # B
                    colorIndex = 0 
                    bx = (center[0])+97
                    x1 = (center[0])
                    color3[0]=bx
                elif 162 <= center[0] <= 318: # G
                    colorIndex = 1 
                    gy = (center[0])+97-160
                    x2 = (center[0])
                    color3[1]=gy
                elif 322 <= center[0] <= 478: # R
                    colorIndex = 2 
                    rz = (center[0])+97-320
                    x3 = (center[0])
                    color3[2]=rz
                elif 482 <= center[0] <= 638: # Y
                    colorIndex = 3 
                    yy = yz = (center[0])+97-480
                    x4 = (center[0])
                    color3[3]=yy

        #Lower section           
        elif center[1] <= 470 and center[1] >= 380:        
            if (490 <= center[0] <= 640):
                if (flagc==0):
                    flagc=1
                    nowc1 = datetime.datetime.now() 
                    nowc2 = datetime.datetime.now()                   
                
                cv2.putText(frame, "{0}".format(3-(nowc2.second-nowc1.second) ), (300, 450), cv2.FONT_HERSHEY_SIMPLEX, 2, (0, 255, 255), 1, cv2.LINE_AA)

                if (490 <= center[0] <= 640 and nowc2.second-3 > nowc1.second and flagc==1 and flagc_up==0): # Clear All
                    cv2.putText(frame, "CLEAR", (500, 450), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 255, 255), 4, cv2.LINE_AA)
                    cv2.putText(paintWindow, "CLEAR", (500, 450), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 255, 255), 4, cv2.LINE_AA)
                    #
                    bpoints = [deque(maxlen=512)]
                    gpoints = [deque(maxlen=512)]
                    rpoints = [deque(maxlen=512)]
                    ypoints = [deque(maxlen=512)]
                    #
                    bindex = 0
                    gindex = 0
                    rindex = 0
                    yindex = 0
                    #
                    paintWindow[67:,:,:] = 255
                    flagc=0
                elif (490 <= center[0] <= 640 and nowc2.second >=58 and flagc==1 and flagc_up==0):
                    sec58up = nowc1.second - 58
                    flagc_up=1
                elif(490 <= center[0] <= 640 and nowc2.second-3 > sec58up and flagc==1 and flagc_up==1):
                    cv2.putText(frame, "CLEAR", (500, 450), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 255, 255), 4, cv2.LINE_AA)
                    cv2.putText(paintWindow, "CLEAR", (500, 450), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 255, 255), 4, cv2.LINE_AA)
                    #
                    bpoints = [deque(maxlen=512)]
                    gpoints = [deque(maxlen=512)]
                    rpoints = [deque(maxlen=512)]
                    ypoints = [deque(maxlen=512)]
                    #
                    bindex = 0
                    gindex = 0
                    rindex = 0
                    yindex = 0
                    #
                    paintWindow[67:,:,:] = 255
                    flagc=0
                    flagc_up=0
                    sec58up=100
                else:
                    nowc2 = datetime.datetime.now()
            else:
                nowc1 = datetime.datetime.now() 
                nowc2 = datetime.datetime.now()
                flagc=0

        else :
            if colorIndex == 0: # Blue
                bpoints[bindex].appendleft(center)
            elif colorIndex == 1: # Green
                gpoints[gindex].appendleft(center)
            elif colorIndex == 2: # Red
                rpoints[rindex].appendleft(center)
            elif colorIndex == 3: # Yellow
                ypoints[yindex].appendleft(center)
  
        
        if (int(radius)<=contour_scale):
            bpoints.append(deque(maxlen=512))
            bindex += 1
            gpoints.append(deque(maxlen=512))
            gindex += 1
            rpoints.append(deque(maxlen=512))
            rindex += 1
            ypoints.append(deque(maxlen=512))
            yindex += 1

            
    # Append the next deque when no contours are detected
    else:
        bpoints.append(deque(maxlen=512))
        bindex += 1
        gpoints.append(deque(maxlen=512))
        gindex += 1
        rpoints.append(deque(maxlen=512))
        rindex += 1
        ypoints.append(deque(maxlen=512))
        yindex += 1

  
    # Draw lines (Blue, Green, Red and Yellow)
    points = [bpoints, gpoints, rpoints, ypoints]
    for i in range(len(points)):
        for j in range(len(points[i])):
            for k in range(1, len(points[i][j])):
                if points[i][j][k - 1] is None or points[i][j][k] is None:
                    continue
                cv2.line(frame, points[i][j][k - 1], points[i][j][k], (255,255,255), 12)
                cv2.line(frame, points[i][j][k - 1], points[i][j][k], colors[i], 11)
                cv2.line(paintWindow, points[i][j][k - 1], points[i][j][k], (255,255,255), 12)
                cv2.line(paintWindow, points[i][j][k - 1], points[i][j][k], colors[i], 11)

                cv2.line(paintWindow, points[i][j][k - 1], points[i][j][k],(255,255,255), 1)
                cv2.line(frame, points[i][j][k - 1], points[i][j][k],(255,255,255), 1)

                
    # Get the moments to calculate the center of the contour (in this case Circle)
    if len(cnts) > 0:
        M = cv2.moments(cnt)
        center = (int(M['m10'] / M['m00']), int(M['m01'] / M['m00']))
            
        if center[1] <= 470 and center[1] >= 380:
                
            if 0 <= center[0] <= 150: # Save paint
                now2 = datetime.datetime.now()
                if (flagt==0 or now2.second >= now1.second+1):
                    stas = cv2.imwrite("img{0}.jpg".format(lastIndex),paintWindow[100:size_pw[0]-100,:] )
                    print("{0}: {1}".format(lastIndex, stas))
                    cv2.putText(frame, "SAVE", (10, 450), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 255, 255), 4, cv2.LINE_AA)
                    cv2.putText(paintWindow, "SAVE", (10, 450), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 255, 255), 4, cv2.LINE_AA)
                    if (stas):
                        flagt=1
                        now1 = datetime.datetime.now()
                        lastIndex += 1             
                    
                    
    # Show the frame and the paintWindow image
    cv2.imshow("Camera Paint", frame)
    cv2.imshow("White Paint", paintWindow)
    
    
    size_pw=paintWindow.shape
    
    
    ##Keyboard Shortcuts ('Esc' , 'q' , 'c' , 's'  ###  'b' , 'g' , 'r' , 'y')
    
    # If the 'Esc' key is pressed, stop the loop
    if cv2.waitKey(1) == 27:
        break
    # If the 'q' key is pressed, stop the loop
    if cv2.waitKey(1) == ord("q"):
        break
    # If the 'c' key is pressed, clear all    
    elif cv2.waitKey(1) == ord("c"):
        cv2.putText(frame, "CLEAR", (500, 450), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 255, 255), 4, cv2.LINE_AA)
        cv2.putText(paintWindow, "CLEAR", (500, 450), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 255, 255), 4, cv2.LINE_AA)
        #
        bpoints = [deque(maxlen=512)]
        gpoints = [deque(maxlen=512)]
        rpoints = [deque(maxlen=512)]
        ypoints = [deque(maxlen=512)]
        #
        bindex = 0
        gindex = 0
        rindex = 0
        yindex = 0
        #
        paintWindow[67:,:,:] = 255
        
    elif cv2.waitKey(1) == ord("b"):
        colorIndex = 0
    elif cv2.waitKey(1) == ord("g"):
        colorIndex = 1
    elif cv2.waitKey(1) == ord("r"):
        colorIndex = 2
    elif cv2.waitKey(1) == ord("y"):
        colorIndex = 3    
        
    # If the 's' key is pressed, save the paint     
    elif cv2.waitKey(1) == ord("s"):
        stas = cv2.imwrite("img{0}.jpg".format(lastIndex),paintWindow[100:size_pw[0]-100,:] )
        print("{0}: {1}".format(lastIndex, stas))
        if (stas):
            cv2.putText(frame, "SAVE", (10, 450), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 255, 255), 4, cv2.LINE_AA)
            cv2.putText(paintWindow, "SAVE", (10, 450), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 255, 255), 4, cv2.LINE_AA)
            lastIndex += 1   
        
    
# Cleanup the camera and close any open windows
camera.release()
cv2.destroyAllWindows()
```
